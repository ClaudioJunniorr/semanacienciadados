### Separando por ;

```python
import pandas as pd
# Importando os dados e os inserindo em uma variável
url="dataset/aluguel.csv"
# O método read csv recebe url como parametro para ler.
# url, é um parametro e sep é outro parametro.
dados=pd.read_csv(url,sep=";")
# O sep= no parâmetro, ele usa para separar e organizar com ;
# Iremos implementar em como poder ver onde quiser nos dados.
```
